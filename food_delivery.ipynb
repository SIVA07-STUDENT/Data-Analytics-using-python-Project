```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("food_delivery.csv")

print(df.head())

# Total Orders
print("Total Orders:", len(df))

# Most Popular Food
popular = df["Item"].value_counts()
print(popular)

# Location Wise Orders
location = df["Location"].value_counts()
print(location)

# Peak Order Time
df["Hour"] = pd.to_datetime(df["Time"]).dt.hour
peak = df["Hour"].value_counts().sort_index()
print(peak)

# Total Sales
sales = df["Price"].sum()
print("Total Sales =", sales)

# Popular Food Chart
popular.plot(kind="bar")
plt.title("Most Popular Food")
plt.xlabel("Food Item")
plt.ylabel("Orders")
plt.show()

# Peak Order Time Chart
peak.plot(kind="line", marker="o")
plt.title("Peak Order Time")
plt.xlabel("Hour")
plt.ylabel("Orders")
plt.show()

# Location Orders Chart
location.plot(kind="pie", autopct="%1.1f%%")
plt.title("Location Wise Orders")
plt.ylabel("")
plt.show()
```
